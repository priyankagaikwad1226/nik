#### UpSampling using RandomOverSampler

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
random_os = RandomOverSampler(sampling_strategy=0.70)
```

```python
X = df.drop('Class',axis=1)
```

```python
y = df['Class']
```

```python
X_resample,y_resample = random_os.fit_resample(X,y)
```

```python
y_resample.value_counts()
```

```python
X_resample
```

#### UnderSampling using RandomUnderSampler

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
random_us = RandomUnderSampler(sampling_strategy=0.60)
```

```python
X_RUSresample,y_RUSresample = random_us.fit_resample(X,y)
```

```python
y_RUSresample.value_counts()
```

```python
X_RUSresample
```

#### UnderSampling using NearMiss

```python
from imblearn.under_sampling import NearMiss
```

```python
X_NM_resample_under,y_NM_resample_under = under_rs.fit_resample(X,y)
```

```python
y_NM_resample_under.value_counts()
```

#### UpSampling using SMOTETEK (Synthetic Minority Over-sampling Technique)

```python
import pandas as pd
```

```python
df=pd.read_csv('creditcard.csv')
```

```python
X = df.drop('Class',axis=1)
```

```python
y = df['Class']
```

```python
X
```

```python
from imblearn.combine import SMOTETomek
```

```python
obj = SMOTETomek(sampling_strategy=0.60)
```

```python
x_resample_smote,y_resample_smote = obj.fit_resample(X,y)
```

```python
y_resample_smote.value_counts()
```
